```python
import folium

# Find the center of all the locations and prepare the folium map

center_lat=df2.Latitude.mean()
center_long=df2.Longitude.mean()

```

```python
venues_map = folium.Map(location=[center_lat, center_long], zoom_start=13)

# add markers to map
for lat, lng, borough, neighe in zip(df2['Latitude'], df2['Longitude'], df2['Borough'], df2['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(venues_map)  
    
venues_map
```
